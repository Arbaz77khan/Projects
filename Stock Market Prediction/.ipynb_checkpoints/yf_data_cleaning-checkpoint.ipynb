```python
import pandas as pd
```

```python
data = pd.read_csv('stock_data/SUZLON.NS_2024-01-01_to_2024-11-09.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
data.sample()
```

```python
data.isnull()
```

```python
data.rename(columns={'Price': 'Date'}, inplace=True)
```

```python
data.sample()
```

```python
data.drop(index=[0,1], inplace=True)
```

```python
data.head()
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date']).dt.tz_localize(None)
```

```python
text_to_num_cols = ['Adj Close', 'Close', 'High', 'Low', 'Open']
data[text_to_num_cols] = data[text_to_num_cols].apply(lambda x: pd.to_numeric(x, errors='coerce').round(2))
data['Volume'] = pd.to_numeric(data['Volume'], errors='coerce')
```
